```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Type'].value_counts().plot(kind='bar')
plt.title('Count of Post Types')
plt.xlabel('Post Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(df['Total Interactions'], bins=200)
plt.title('Distribution of Total Interactions')
plt.xlabel('Total Interactions')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(df['Lifetime Post Total Reach'], df['Total Interactions'])
plt.title('Reach vs Total Interactions')
plt.xlabel('Total Reach')
plt.ylabel('Total Interactions')
plt.show()
```

```python
df['Paid'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Paid vs Unpaid Posts')
plt.ylabel('')
plt.show()
```

```python
monthly_avg = df.groupby('Post Month')['Total Interactions'].mean()
monthly_avg.plot(kind='line')
plt.title('Average Interactions per Month')
plt.xlabel('Month')
plt.ylabel('Average Interactions')
plt.show()
```

```python
df['Engagement_Rate'] = df['Total Interactions'] / df['Lifetime Post Total Reach']

plt.hist(df['Engagement_Rate'], bins=20)
plt.title('Engagement Rate Distribution')
plt.show()
```
